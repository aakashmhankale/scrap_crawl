```python
import requests
from bs4 import BeautifulSoup
r=requests.get("http://www.century21.com/real-estate/irvine-ca/LCCAIRVINE/")
c=r.content
soup=BeautifulSoup(c,"html.parser")
all=soup.find_all("div",{"class":"property-card-primary-info"})
all[0].find("a",{"class":"listing-price"}).text.replace("\n","").replace(" ","")
```

```python
l=[]
base_url="http://www.century21.com/real-estate/irvine-ca/LCCAIRVINE/?k=1&p="
for page in range(0,10):
    #print(base_url+str(page))
    r=requests.get(base_url+str(page))
    c=r.content
    soup=BeautifulSoup(c,"html.parser")
    all=soup.find_all("div",{"class":"property-card-primary-info"})
    for item in all:
        d={}
        d["price"]=item.find("a",{"class":"listing-price"}).text.replace("\n","").replace(" ","")
        d["address"]=item.find("div",{"class":"property-address"}).text.replace("\n","").replace(" ","")
        d["city"]=item.find("div",{"class":"property-city"}).text.replace("\n","").replace(" ","")
        try:
            d["bed"]=item.find("div",{"class":"property-beds"}).text.replace("\n","").replace(" ","")
        except:
            d["bed"]=None
        try:
            d["bath"]=item.find("div",{"class":"property-baths"}).text.replace("\n","").replace(" ","")
        except:
            d["bath"]=None
        try:
            d["Area"]=item.find("div",{"class":"property-sqft"}).text.replace("\n","").replace(" ","")
        except:
            d["Area"]=None
        print("\n")
        l.append(d)
l
```

```python
import pandas
final=pandas.DataFrame(l)
final
final.to_csv("Property_scapping.csv")
```
